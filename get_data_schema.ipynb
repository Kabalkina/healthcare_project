```python
# Mapping pandas dtypes to SQL types
dtype_mapping = {
    'object': 'text',
    'int64': 'integer',
    'float64': 'double precision',
    'bool': 'boolean',
    'datetime64[ns]': 'timestamp'
}
```

```python
# Build SQL columns
columns_sql = []
for col, dtype in df.dtypes.items():
    sql_type = dtype_mapping.get(str(dtype), 'text')  # Default to text if unknown
    columns_sql.append(f"    {col} {sql_type}")
```

```python
# Mapping pandas dtypes to BigQuery SQL types
dtype_mapping_bq = {
    'object': 'STRING',
    'int64': 'INTEGER',
    'float64': 'NUMERIC',
    'bool': 'BOOLEAN',
    'datetime64[ns]': 'TIMESTAMP'
}
```

```python
# Generate field definitions (no description)
columns_sql = []
for col, dtype in df.dtypes.items():
    sql_type = dtype_mapping_bq.get(str(dtype), 'STRING')  # Default to STRING if unknown
    field_def = f"{col} {sql_type}"
    columns_sql.append(field_def)
```
